## Load the Feature-Engineered Dataset

## Dataset Overview

## Statistical Outlier Detection using Z-Score

Statistical outliers are observations that are significantly different from the majority of the data. Detecting outliers helps improve data quality and prevents unusual values from negatively affecting future analysis and Machine Learning models.

In this section, the Z-Score method is used to identify values that are more than three standard deviations away from the mean.

```python
# Select continuous numerical columns only
numeric_columns = ["Age", "Fare"]

print("Numerical Columns:")
print(numeric_columns)
```

```python
# Calculate Z-Scores
z_scores = np.abs(zscore(df[numeric_columns], nan_policy="omit"))

# Replace NaN values with 0 so they are not treated as outliers
z_scores = np.nan_to_num(z_scores)

print("Z-Score matrix created successfully!")
```

```python
# Create Boolean mask
zscore_mask = (z_scores < 3).all(axis=1)

# Filter dataset
zscore_df = df[zscore_mask]

print("Original Shape:", df.shape)
print("After Z-Score:", zscore_df.shape)
```

### Observation

The Z-Score method identifies values that are more than three standard deviations away from the mean. It works best when the numerical data is approximately normally distributed. Records identified as statistical outliers were removed using vectorized Boolean indexing.

## Statistical Outlier Detection using IQR

The Interquartile Range (IQR) method detects outliers using the spread of the middle 50% of the data. It is more robust than the Z-Score method for skewed distributions because it relies on quartiles rather than the mean and standard deviation.

```python
# Calculate Quartiles
Q1 = zscore_df[numeric_columns].quantile(0.25)
Q3 = zscore_df[numeric_columns].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

print("IQR calculated successfully!")
```

```python
# Calculate Lower and Upper Bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower and Upper Bounds calculated.")
```

```python
# Create Boolean mask
iqr_mask = ~(
    (zscore_df[numeric_columns] < lower_bound) |
    (zscore_df[numeric_columns] > upper_bound)
).any(axis=1)

# Apply mask
final_df = zscore_df[iqr_mask]

print("Original Shape:", df.shape)
print("After Z-Score:", zscore_df.shape)
print("After IQR:", final_df.shape)
print("Total Rows Removed:", df.shape[0] - final_df.shape[0])
```

### Observation

The IQR method uses the first and third quartiles to identify extreme values. Unlike the Z-Score method, it is less affected by skewed distributions and is often more reliable for real-world datasets that are not normally distributed.
