## Data Cleaning

```python
print("Missing Values:")
df.isnull().sum()
```

```python
# Fill missing values in numerical columns with the median
numeric_columns = df.select_dtypes(include=np.number).columns

df[numeric_columns] = df[numeric_columns].fillna(
    df[numeric_columns].median()
)

# Fill missing values in categorical columns with the mode
categorical_columns = df.select_dtypes(exclude=np.number).columns

for column in categorical_columns:
    df[column] = df[column].fillna(df[column].mode()[0])

print("Missing values handled successfully!")
```

```python
print("Remaining Missing Values:")
df.isnull().sum()
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
df = df.drop_duplicates()

print("Dataset Shape After Removing Duplicates:")
print(df.shape)
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

## Export Cleaned Dataset

```python
from pathlib import Path

output_path = Path.cwd() / "cleaned_dataset.csv"

df.to_csv(output_path, index=False)

print("✅ Cleaned dataset exported successfully!")
print("Saved to:", output_path)
```

# Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is the process of understanding a dataset using statistical summaries and visualizations before applying Machine Learning models.

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

## Business Insight 1

The correlation heatmap shows the relationships between numerical variables in the dataset. Strong positive or negative correlations help identify which variables have the greatest influence on business performance and can support future decision-making.

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Revenue"], bins=30, kde=True)

plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Count")

plt.show()
```

## Business Insight 2

The revenue distribution indicates how sales are spread across orders. Most orders fall within a common revenue range, while a smaller number of high-value orders contribute significantly to total revenue.

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Profit"], bins=30, kde=True)

plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Count")

plt.show()
```

## Business Insight 3

The profit distribution helps identify whether the business consistently generates profits or experiences losses on some transactions. Understanding this distribution can help optimize pricing and cost management strategies.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Revenue"])

plt.title("Revenue Box Plot")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x="Category", data=df)

plt.xticks(rotation=45)

plt.title("Number of Products by Category")

plt.show()
```

```python
plt.figure(figsize=(10,5))

df.groupby("Category")["Revenue"].sum().plot(kind="bar")

plt.title("Total Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")

plt.show()
```

## Business Insight 4

The category-wise revenue comparison highlights which product categories generate the highest sales. High-performing categories can be prioritized for marketing and inventory planning, while lower-performing categories may require improvement strategies.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="Revenue",
    y="Profit",
    data=df
)

plt.title("Revenue vs Profit")

plt.show()
```

## Business Insight 5

The scatter plot between revenue and profit shows how profitability changes with increasing revenue. Transactions with high revenue but relatively low profit may indicate high operational costs or excessive discounts.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Shipping_Days"])

plt.title("Shipping Days Distribution")

plt.show()
```
