# Selecting Features (X) and Target Variable (y)

In supervised machine learning, the dataset is divided into:

- **X (Features):** Independent variables used to make predictions.
- **y (Target):** The dependent variable that the model learns to predict.

For this project, the target variable is **Survived**, where:

- 0 = Did not survive
- 1 = Survived

```python
# Target variable
y = df["Survived"]

print("Target variable created successfully!")
```

```python
# Select useful numerical features
X = df[
    [
        "Pclass",
        "Age",
        "SibSp",
        "Parch",
        "Fare",
        "Sex_Male",
        "Embarked_Q",
        "Embarked_S"
    ]
]

print("Feature matrix created successfully!")
```

```python
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

# Train-Test Split

Before training a Machine Learning model, the dataset is divided into:

- **Training Set (80%)** → Used to teach the model.
- **Testing Set (20%)** → Used to evaluate the model on unseen data.

Using separate training and testing data helps prevent **data leakage** and provides a more reliable evaluation of the model.

```python
from sklearn.model_selection import train_test_split

print("train_test_split imported successfully!")
```

```python
# Split the dataset into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Dataset split successfully!")
```

```python
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)

print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

print("Logistic Regression imported successfully!")
```

```python
# Initialize the Logistic Regression model

model = LogisticRegression(random_state=42)

print("Model created successfully!")
```

```python
# Train the model using training data

model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
# Predict on unseen test data

y_pred = model.predict(X_test)

print("Predictions generated successfully!")
```

```python
print("First 10 Predictions:")
print(y_pred[:10])
```

```python
comparison = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

comparison.head(10)
```

# Model Evaluation

After training the Logistic Regression model, we evaluate its performance using the testing dataset.

The accuracy score measures the percentage of correct predictions made by the model on unseen data.

```python
from sklearn.metrics import accuracy_score

print("Accuracy Score imported successfully!")
```

```python
# Calculate model accuracy

accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)
```

```python
print(f"Accuracy: {accuracy * 100:.2f}%")
```

## Observation

The Logistic Regression model was evaluated using the testing dataset. The accuracy score indicates the percentage of correct predictions made on unseen data.

A higher accuracy generally indicates better predictive performance, but accuracy should be interpreted together with other evaluation metrics depending on the problem.
